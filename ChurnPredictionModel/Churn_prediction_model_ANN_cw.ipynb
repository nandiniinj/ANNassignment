```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
train=pd.read_csv('churn_train.csv')
```

```python
train.head()
```

```python
test=pd.read_csv('churn_test.csv')
```

```python
test.head()
```

```python
Exited=train.Exited
```

```python
Exited.head()
```

```python
train.drop('Exited',axis=1,inplace=True)
test.drop('Exited',axis=1,inplace=True)
```

```python
data=train.append(test)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.EstimatedSalary.dtypes
```

```python
data.Gender=data.Gender.map({'Male':1,'Female':0})
data.Geography=data.Geography.map({'France':1,'Spain':2,'Germany':3})
```

```python
data.head()
```

```python
data.tail()
```

```python
data.drop('Surname',axis=1,inplace=True)
```

```python
data.head()
```

```python
train_X=data.iloc[:800,]
train_y=Exited
X_test=data.iloc[500:,]
```

```python
train_X,test_X,train_y,test_y=train_test_split(train_X,train_y,test_size=0.2,random_state=0)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model1=Perceptron(eta0=1.0,max_iter=1000,tol=1e-3,random_state=10)
model1.fit(train_X,train_y)
```

```python
prediction1=model1.predict(test_X)
```

```python
prediction1
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(test_y,prediction1)
```

```python
accuracy_score(test_y,prediction1)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision_score(test_y,prediction1,average='binary')
```

```python
recall_score(test_y,prediction1,average='binary') 
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(test_y,prediction1,average='binary')
```

```python
df_output=pd.DataFrame()
```

```python
outp=model1.predict(X_test)
df_output['status']=outp
```

```python
df_output.head()
```

```python
df_output[['status']].to_csv(r'output.csv',index=False)
```
