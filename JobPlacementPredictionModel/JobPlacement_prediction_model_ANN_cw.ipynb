```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
pla_train=pd.read_csv('placement_train.csv')
```

```python
pla_train.head()
```

```python
pla_test=pd.read_csv('placement_test.csv')
```

```python
pla_test.head()
```

```python
pla_train.status=pla_train.status.map({'Placed':1,'Not Placed':0})
pla_train.head()
```

```python
status=pla_train.status
```

```python
status
```

```python
pla_train.drop('status',axis=1,inplace=True)
pla_test.drop('status',axis=1,inplace=True)
```

```python
data=pla_train.append(pla_test)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.degree_percentage.dtypes
```

```python
data.gender=data.gender.map({'M':1,'F':0})
data.ssc_board=data.ssc_board.map({'Central':1,'Others':0})
data.hsc_board=data.hsc_board.map({'Central':1,'Others':0})
data.hsc_subject=data.hsc_subject.map({'Commerce':1,'Arts':2,'Science':3})
data.undergrad_degree=data.undergrad_degree.map({'Others':1,'Comm&Mgmt':2,'Sci&Tech':3})
data.work_experience=data.work_experience.map({'Yes':1,'No':0})
data.specialisation=data.specialisation.map({'Mkt&Fin':1,'Mkt&HR':0})
```

```python
data.head()
```

```python
data.tail()
```

```python
train_X=data.iloc[:172,]
train_y=status
X_test=data.iloc[172:,]
seed=10
```

```python
train_X,test_X,train_y,test_y=train_test_split(train_X,train_y,random_state=seed)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model1=Perceptron(eta0=1.0,max_iter=1000,tol=1e-3,random_state=10)
model1.fit(train_X,train_y)
```

```python
prediction1=model1.predict(test_X)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(test_y,prediction1)
```

```python
accuracy_score(test_y,prediction1)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision_score(test_y,prediction1,average='binary')
```

```python
recall_score(test_y,prediction1,average='binary') 
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(test_y,prediction1,average='binary')
```

```python
df_output=pd.DataFrame()
outp=model1.predict(X_test)
df_output['status']=outp
```

```python
df_output.head()
```

```python
df_output.tail()
```

```python
df_output[['status']].to_csv(r'output.csv',index=False)
```
