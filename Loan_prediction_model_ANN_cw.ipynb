```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
loan_train=pd.read_csv('train.csv')
```

```python
loan_train.head()
```

```python
loan_train.tail()
```

```python
loan_test=pd.read_csv('test.csv')
```

```python
loan_test.head()
loan_test.tail()
```

```python
loan_test.head()
```

```python
loan_train.Loan_Status=loan_train.Loan_Status.map({'Y':1,'N':0})
loan_train.head()
```

```python
Loan_Status=loan_train.Loan_Status
```

```python
Loan_Status
```

```python
loan_train.drop('Loan_Status',axis=1,inplace=True)
```

```python
Loan_ID=loan_test.Loan_ID
```

```python
data=loan_train.append(loan_test)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.Dependents.dtypes
```

```python
data.Gender=data.Gender.map({'Male':1,'Female':0})
data.Married=data.Married.map({'Yes':1,'No':0})
data.Dependents=data.Dependents.map({'0':0,'1':1,'2':2,'3+':3})
data.Education=data.Education.map({'Graduate':1,'Not Graduate':0})
data.Self_Employed=data.Self_Employed.map({'Yes':1,'No':0})
data.Property_Area=data.Property_Area.map({'Urban':2,'Rural':0,'Semiurban':1})
```

```python
data.head()
```

```python
data.tail()
```

```python
data.Credit_History.fillna(np.random.randint(0,2),inplace=True)
data.Married.fillna(np.random.randint(0,2),inplace=True)
data.Gender.fillna(np.random.randint(0,2),inplace=True)
data.Dependents.fillna(data.Dependents.median(),inplace=True)
data.Self_Employed.fillna(np.random.randint(0,2),inplace=True)
data.Loan_Amount_Term.fillna(data.Loan_Amount_Term.mean(),inplace=True)
data.LoanAmount.fillna(data.LoanAmount.median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop('Loan_ID',inplace=True,axis=1)
```

```python
loan_train_X=data.iloc[:614,]
loan_train_y=Loan_Status
X_test=data.iloc[614:,]
seed=100
```

```python
loan_train_X,loan_test_X,loan_train_y,loan_test_y=train_test_split(loan_train_X,loan_train_y,random_state=seed)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model1=Perceptron(eta0=1.0,max_iter=1000,tol=1e-3,random_state=42)
model1.fit(loan_train_X,loan_train_y)
```

```python
prediction1=model1.predict(loan_test_X)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(loan_test_y,prediction1)
```

```python
accuracy_score(loan_test_y,prediction1)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision_score(loan_test_y,prediction1,average='binary')
```

```python
recall_score(loan_test_y,prediction1,average='binary')
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(loan_test_y,prediction1,average='binary')
```
